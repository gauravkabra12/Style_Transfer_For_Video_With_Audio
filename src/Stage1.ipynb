```
import cv2
import numpy as np
import os
cap = cv2.VideoCapture('./storm.mp4')

currentFrame = 0
try:
  while True:
      ret, frame = cap.read()
      if not ret:
        break
      name = './Frames/' + str(currentFrame) + '.jpg'
      print ('Creating...' + name)
      cv2.imwrite(name, frame)
      currentFrame += 1
except:
  pass
  
cap.release()
cv2.destroyAllWindows()
```
